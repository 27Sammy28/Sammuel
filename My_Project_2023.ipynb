###  In contrast to pure programming, in mathematics we mostly deal with indeterminatesin order to solve equations or provide general results. In Sage, these indeterminates are realized by so-called symbolic variables, which have to be distinguished from Python variables very carefully. 

```python
sage: z= SR.var('x')
sage: 2*z  -1    
```

#### To clean such expressions up, Sage provides several simplification methods. The most elementary method is given by simplify.

```python
sage: x = var('x')
sage: cos(x)^2 + sin(x)^2
```

```python
sage: x= var('x')
sage: (x^2/x).simplify()    
```

###  For more complicated expressions, the type of simplification has to be made explicit.Expressions containing factorials can be simplified with simplify_factorial:

```python
sage: n= var('n')
sage: expr= factorial(n+1) / factorial(n)
sage: expr.simplify_factorial()    
```

### Similarly, simplify_rational simplifies expressions containing fractions:

```python
sage: x= var('x')
sage: expr= (x^2-1) /(x-1) 
sage: expr.simplify_rational()   
```

```python
sage: x= var('x')
sage: expr= (x^6-1) /(x+1) 
sage: expr.simplify_rational()   
```

### To deal with expressions such as square roots, logarithms or exponentials, we use  canonicalize_radical:

```python
sage: x, y = var('x,y')
sage: (sqrt(abs(x)^2)).canonicalize_radical()
```

### The command canonicalize_radical is special among the simplification methods. In general, roots and logarithms have different branches in the complex plane. For 2 instance, the expression x is not uniquely defined in C. Namely, we can have x or x depending on the branch we have picked. The method canonicalize_radical chooses one branch by heuristics and simplifies on that.

# Sequences and Series

### Sage is a good tool to study the behavior of sequences and series. For example, limits of sequences or functions can be easily computed with the limit method.

```python
sage: x= var('x')
sage: limit ((x^(1/3)-2) / ((x+19)^(1/3)-3), x=8)       
```

### As we can see, limit has two arguments. The first is the function and the second is thepoint at which we want to take the limit. Instead of limit we can use its alias lim. The limit functions calculates by default the limit to the left. We can add the dir option asa third argument to specify the direction of the limit. To calculate limits to the left wechoose the option minus, respectively the option plus to calculate the limit to the right.

```python
sage: n = var('n'); u(n) = n^100 / 100^n
sage: u(1.)
```

```python
sage: u(2.)
```

```python
sage: u(3.)
```

```python
sage: u(4.)
```

```python
sage: u(5.)
```

```python
sage: plot(u(n), n, 1, 40)
```

```python
sage: lim(u(n),n= -infinity)
```

```python
sage: lim(u(n),n= infinity)
```

```python
sage: ((1+arctan(x))^(1/x)).series(x == 0, 3)
```

```python
sage: ((1-arctan(x))^(1/x)).series(x == 0, 3)
```

####  Here, Sage displays that there are terms of higher order. To extract the regular part of a power series expansion obtained by the above method, we use the truncate method.

```python
sage: ((1+arctan(x))^(1/x)).series(x == 0, 3).truncate()
```

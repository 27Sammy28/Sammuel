```python
import numpy as np
import pylab
from math import pi
```

```python
x = np.linspace(0, 10, 100)
y=np.sin(x)
z=np.cos(x)
pylab.figure(figsize=(14,8))
pylab.plot(x,y,'b',label='sin(x)',linewidth=4)
pylab.plot(x,z,'r',label='cos(x)',linewidth=4)

pylab.xlabel('X-axis',fontsize=25)
pylab.ylabel('sin and cosx',fontsize=25)
pylab.title('sine and cosine function ',fontsize=20)
pylab.legend(loc=1,fontsize=17)
pylab.rc('xtick',labelsize=22)
pylab.rc('ytick',labelsize=22)
pylab.show()
```

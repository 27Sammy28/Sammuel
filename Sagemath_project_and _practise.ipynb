#  Algorithmics

```python
sage: def fct(x,y):
        return  (x^2 + y^2)
sage: fct(3,4)
```

```python
sage: def means(x,y):
        a= (x+y)/2
        g = sqrt(x * y)
        h = 2 * (1/x + 1/y)^(-1)
        return a,g,h
sage: means(6, 8)
```

```python
sage: amean = lambda x, y: (x+y) / 2
sage: gmean = lambda x, y: sqrt(x * y)
sage: hmean = lambda x, y: 2 * (1/x + 1/y)^(-1)
sage: amean(6, 8), gmean(6, 8), hmean(6, 8)
```
